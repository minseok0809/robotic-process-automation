### Development Environment

```python
import cv2
import time
import webbrowser
from selenium import webdriver
import matplotlib
import matplotlib.pyplot as plt
```

### Meteorological Administration Forecast

```python
wd = webdriver.Chrome()
url = 'https://www.weather.go.kr/w/weather/forecast/short-term.do'
wd.get(url)
time.sleep(3)
wd.save_screenshot('forecast.png')
wd.quit()
img1_path = 'forecast.png'
img1_orig = cv2.imread(img1_path) 

hotrizontal_ratio = img1_orig.shape[0] / 50
vertical_ratio = img1_orig.shape[1] / 50

plt.figure(figsize=(hotrizontal_ratio, vertical_ratio))
plt.imshow(cv2.cvtColor(img1_orig, cv2.COLOR_BGR2RGB), interpolation='nearest', aspect='auto')
ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
plt.show()
```
